```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2024-01-01", end="2025-01-01")
data.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data["Close"])
plt.title("Apple Stock Price (2024)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
data["Daily Return"] = data["Close"].pct_change()
data["Daily Return"].head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data["Daily Return"])
plt.title("Daily Returns of Apple Stock")
plt.xlabel("Date")
plt.ylabel("Return")
plt.show()
```

```python
data["MA20"] = data["Close"].rolling(window=20).mean()

plt.figure(figsize=(10,5))
plt.plot(data["Close"], label="Close Price")
plt.plot(data["MA20"], label="20-Day Moving Average")
plt.title("Apple Stock Price with Moving Average")
plt.legend()
plt.show()
```
