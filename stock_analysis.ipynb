## Data Collection

We use the `yfinance` library to download historical stock price data for Apple (AAPL). The dataset includes Open, High, Low, Close, Volume, and Adjusted Close prices.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2024-01-01", end="2025-01-01")
data.head()
```

## Data Processing
We compute derived features:
- Daily Returns (percentage change in closing price)
- 20-day Moving Average (trend smoothing)
- Rolling Volatility (risk estimation over time)

```python
data["Daily Return"] = data["Close"].pct_change()
data["MA20"] = data["Close"].rolling(window=20).mean()
data["Rolling Volatility"] = data["Daily Return"].rolling(window=20).std()
```

## Statistical Summary
We compute descriptive statistics to understand the distribution of stock prices and returns.

```python
data.describe()
```

## Analysis and Visualization
We visualize price trends, returns, and volatility to understand market behavior.

```python
plt.figure(figsize=(10,5))
plt.plot(data["Close"], label="Close Price")
plt.plot(data["MA20"], label="20-Day Moving Average")
plt.title("AAPL Price Trend with Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data["Daily Return"])
plt.title("Daily Returns of AAPL")
plt.xlabel("Date")
plt.ylabel("Return")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data["Rolling Volatility"])
plt.title("20-Day Rolling Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.show()
```

## Insights and Interpretation

- Stock prices show short-term fluctuations with an overall trend structure.
- Daily returns fluctuate around zero, reflecting normal market behavior.
- Moving averages help smooth noise and reveal underlying trends.
- Volatility changes over time, indicating dynamic market risk.

### Key Takeaway
Combining price trends, returns, and volatility provides a more complete understanding of financial time-series behavior than price alone.
